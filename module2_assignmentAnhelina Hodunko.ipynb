## 4.1
Using if, elif, and else to compare numbers.

```python
secret = 7
guess = 5

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

## 4.2
Using if/else to match small and green objects.

```python
small = True
green = True

if small:
    if green:
        print('pea')
    else:
        print('cherry')
else:
    if green:
        print('watermelon')
    else:
        print('pumpkin')
```

## 6.1
Using a for loop to print values from a list.

```python
for value in [3, 2, 1, 0]:
    print(value)
```

## 6.2
Using a while loop to compare numbers.

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

## 6.3
Using a for loop with range().

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
